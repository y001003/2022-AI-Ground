# MostPopular

```python
df_merge['watched'] = 1
```

```python
# 실제로 시청된 value_counts
mp_counts25 = df_merge['album_id'].value_counts().head(25)
mp_counts25
```

```python
pivot = pd.pivot_table(df_merge,values='watched',index='profile_id',columns='album_id',aggfunc='sum')
pivot
```

```python
pivot[pivot[16].notnull()]
```

```python
# 가장 많은 사람에게 보여진 영상은 무엇일까?
```

```python
mp = []
for i in pivot.columns:
    
    leng = pivot[pivot[i].notnull()].shape[0]
    if (len(mp) == 0):
        mp.append((i,leng))
    elif leng > mp[0][1]:
        mp.append((i,leng))
        mp.sort(key=lambda x:x[1])
```

```python
# (album_id, 재생된 사람 수)
mp
```

```python
mp_id = [id[0] for id in mp]
mp_id
```

```python
# 많은사람에게 노출되었지만 횟수로 mp_25 들지 못한 영상들
set(mp_id) - set(mp_counts25.index)
```

```python
mp_id = mp_id[::-1]
```

```python
mp_submit = sample.copy()
```

```python
mp_submit['predicted_list'] = sample['predicted_list'].map(lambda x : mp_id)
```
